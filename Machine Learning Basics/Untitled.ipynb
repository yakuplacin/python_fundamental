```python
import pandas as pd
X = pd.read_csv('train.csv', index_col='Id') 
X_test = pd.read_csv('test.csv', index_col='Id')
```

```python
print(X.shape)
```

```python
X.dropna(axis = 0, subset=["SalePrice"], inplace = True)
X
```

```python
print(X.shape)
```

```python
y = X.SalePrice
y
```

```python
X.drop(['SalePrice'], axis=1, inplace=True)
```

```python
X
```

```python
from sklearn.preprocessing import OrdinalEncoder #(5 pts)

object_cols = [col for col in X.columns if X[col].dtype == "object"] #(13 pts+10 pts)

print(object_cols)
print(len(object_cols))

# Columns that can be safely ordinal encoded
good_label_cols = [col for col in object_cols if 
                   set(X_test[col]).issubset(set(X[col]))] #(14 pts+10 pts)
        
# Problematic columns that will be dropped from the dataset
bad_label_cols = list(set(object_cols)-set(good_label_cols)) #(7 pts)

print('Categorical columns that will be ordinal encoded:', good_label_cols)
print('\nCategorical columns that will be dropped from the dataset:', bad_label_cols)

# Drop categorical columns that will not be encoded
X_train_ordinal = X.drop(bad_label_cols, axis=1) #(6 pts+5 pts)
X_test_ordinal = X_test.drop(bad_label_cols, axis=1) #(6 pts+5 pts)

# Apply ordinal encoder 
ordinal_encoder = OrdinalEncoder() #(2 pts)
X_train_ordinal[good_label_cols] = ordinal_encoder.fit_transform(X_train_ordinal[good_label_cols]) #(6 pts+10 pts)
X_test_ordinal[good_label_cols] = ordinal_encoder.transform(X_test_ordinal[good_label_cols]) #(6 pts+10 pts)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer()
```

```python
imputed_X_train = pd.DataFrame(imputer.fit_transform(X_train_ordinal))
imputed_X_test = pd.DataFrame(imputer.transform(X_test_ordinal))

# Fill in the lines below: imputation removed column names; put them back
imputed_X_train.columns = X_train_ordinal.columns
imputed_X_test.columns = X_test_ordinal.columns
```

```python
from xgboost import XGBClassifier
from xgboost import XGBRegressor

model = XGBRegressor(n_estimators=1000, learning_rate=0.05, n_jobs=-1, random_state=0) #(10 pts)
model.fit(imputed_X_train, y) #(4 pts+5 pts+5 pts)
preds_test = model.predict(imputed_X_test) #(4 pts+5 pts)
```

```python

output = pd.DataFrame({'Id': X_test.index,
                       'SalePrice': preds_test})
output.to_csv('submission.csv', index=False)
```
